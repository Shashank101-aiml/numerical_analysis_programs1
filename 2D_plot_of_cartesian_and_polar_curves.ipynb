Plot the scattered curve using the data x = -1,2,3,6,7 and y = 2,4,6,-9,17

```python
import matplotlib.pyplot as plt
x=[-1,2,3,6,7]
y=[2,4,6,-9,17]
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter points')
plt.show()
```

Plot the sine and cosine curve for the range -10 to 10 taking step length as 0.001

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.001)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title("Sine curve and cosine curve")
plt.xlabel("Values of x")
plt.ylabel("Values of sin(x) and cos(x)")
plt.legend()
plt.grid()
plt.show()
```
